```python
import numpy as np
import pandas as pd
a = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
b  = a.T
b
```

# Задание1. Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
fig, ax = plt.subplots()
ax.plot(b[0], b[1], color = 'b', linewidth = 2, label = 'hares')
ax.plot(b[0], b[2],color = 'r', linewidth = 2, label = 'lynxes')
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(10000))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(5000))
plt.xlabel('year', fontsize = 20)
plt.ylabel('amount', fontsize = 20)
fig.set_figwidth(10)
fig.set_figheight(8)
plt.legend()
```

#Задание5. Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
w1 = b[1].mean()
w2 = b[2].mean()
print('Mean hares:', w1)
print('Mean lynxes:',b[1].std())
print('Std hares:', w2)
print('Std lynxes:',b[2].std())
fig, ax = plt.subplots(1, 3)
ax[0].bar(b[0], b[1], label = 'hares')
ax[1].bar(b[0], b[2], color = 'r', label = 'lynxes')
x = np.linspace(1900, 1920, 256)
ax[2].plot(x, x + w1 - x, linewidth = 3, label = 'mean hares')
ax[2].plot(x, x + w2 - x, color = 'r' ,linewidth = 3, label = 'mean lynxes')
ax[0].xaxis.set_major_locator(ticker.MultipleLocator(2))
ax[0].xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax[0].yaxis.set_major_locator(ticker.MultipleLocator(10000))
ax[0].yaxis.set_minor_locator(ticker.MultipleLocator(5000))
ax[1].xaxis.set_major_locator(ticker.MultipleLocator(2))
ax[1].xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax[1].yaxis.set_major_locator(ticker.MultipleLocator(10000))
ax[1].yaxis.set_minor_locator(ticker.MultipleLocator(5000))
ax[0].legend()
plt.ylim([0, 80000])
ax[1].legend()
ax[2].legend()
fig.set_figwidth(40)
fig.set_figheight(10)
```

#Задание6. Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

```python
p = b[1] > b[2]
o = np.array(['H','H','H','H','H','H','H','H','H','H','H','H','H','H','H','H','H','H','H','H','H',])
o1 = np.array(['L','L','L','L','L','L','L','L','L','L','L','L','L','L','L','L','L','L','L','L','L',])
np.copyto(o1, o, where = p)
l = list(o1).count('H')
k = 21 - l
data = np.array([l, k])
labels = ["hares", "lynxes"]
colors = ["b", "r"]
fig = plt.figure(figsize =(10, 7))
plt.pie (data, colors=colors, labels=labels)
```

#Задание9. Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value with the true value and print a ratio of the error to the true value.

```python
#можно использовать модель Хольта Винтерса с сезонностью 10 лет
from statsmodels.tsa.api import ExponentialSmoothing
df = pd.Series(b[1], index = b[0])
train = df['1900': '1920']
test = df[1920]
fit1 = ExponentialSmoothing(train, seasonal_periods=10, trend = 'add', seasonal = 'add').fit()
print('assuming of the population of hares is',fit1.forecast(1).iloc[0])
print('MAPE = ',(100*(test - fit1.forecast(1))/test).iloc[0])
#средняя абсолютная ошибка < 5% 
```
